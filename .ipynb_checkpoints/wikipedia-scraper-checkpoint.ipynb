```python
from requests_html import HTMLSession

session = HTMLSession()

# list of nuclear tests globally
global_url = 'https://en.wikipedia.org/wiki/List_of_nuclear_weapons_tests'
response = session.get(global_url)

global_table = response.html.find('.wikitable', first=True)
data = {}

columns = global_table.find('tr', first=True)
column_labels = [column.text.split('[')[0].rstrip() 
                 for column in columns.find('th')][1:]

rows = global_table.find('tr')
row_labels = []

for i, row in enumerate(rows[1:]):
    country_name = row.find('th')[0].text.split('[')[0].rstrip()
    country_stats = row.find('td')
    if country_name and country_name != 'Totals': 
        row_labels.append(country_name)
        data[country_name] = {}

        for link in row.find('th')[0].absolute_links:
            if 'List_of_nuclear_weapons_tests_of' in link:
                # list of nuclear tests by country
                data[country_name]['url'] = link

        columns_label_index=0
        for i, stat in enumerate(country_stats):
            # ignore reference links
            if i!=1:
                data[country_name][column_labels[columns_label_index]] = stat.text.split(' ')[0].rstrip()
                columns_label_index+=1
            
```

```python
import pprint

pp = pprint.PrettyPrinter(indent=1)
pp.pprint(data)
```

```python
for country, stats in data.items():
    response = session.get(stats['url'])
    
    country_table = response.html.find('.wikitable', first=True)
```
